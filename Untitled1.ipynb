```python
import csv
import datetime

def get_data(filename, target_symbol):
    with open(filename, 'r') as csvfile:
        csvFileReader = csv.reader(csvfile)
        next(csvFileReader)  # skip header

        base_date = None

        for row in csvFileReader:
            symbol = row[7]  # Symbol column

            # ðŸ”¹ Use ONE stock only
            if symbol != target_symbol:
                continue

            date = datetime.datetime.strptime(row[0], "%Y-%m-%d")

            if base_date is None:
                base_date = date

            day_number = (date - base_date).days
            dates.append(day_number)

            prices.append(float(row[4]))  # Close price
```

```python
def predict_prices(dates, prices, x):
    dates = np.reshape(dates, (len(dates), 1))

    svr_lin = SVR(kernel='linear', C=1e3)
    svr_poly = SVR(kernel='poly', C=1e3, degree=2)
    svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)

    svr_lin.fit(dates, prices)
    svr_poly.fit(dates, prices)
    svr_rbf.fit(dates, prices)

    plt.scatter(dates, prices, color='black', label='Data')
    plt.plot(dates, svr_rbf.predict(dates), color='red', label='RBF model')
    plt.plot(dates, svr_lin.predict(dates), color='green', label='Linear model')
    plt.plot(dates, svr_poly.predict(dates), color='blue', label='Polynomial model')
    plt.xlabel('Days')
    plt.ylabel('Price')
    plt.title('Support Vector Regression')
    plt.legend()
    plt.show()

    return (
        svr_rbf.predict(np.array([[x]]))[0],
        svr_lin.predict(np.array([[x]]))[0],
        svr_poly.predict(np.array([[x]]))[0]
    )
```

```python
import pandas as pd
import numpy as np
import datetime

df = pd.read_csv('top10_indian_stocks_3y.csv')

# First row contains stock symbols â†’ save it
symbols = df.iloc[0]

# Remove that row from data
df = df.drop(0).reset_index(drop=True)

# Fix Date column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows where date is missing
df = df.dropna(subset=['Date'])

print(df.head())
```

```python
close_col = 'Close'  # First stock

stock_df = df[['Date', close_col]].dropna()

# Convert date to numeric days
base_date = stock_df['Date'].iloc[0]
stock_df['Day'] = (stock_df['Date'] - base_date).dt.days

dates = stock_df['Day'].values
prices = stock_df[close_col].astype(float).values

print("Loaded rows:", len(dates))
```

```python
from sklearn.svm import SVR
import matplotlib.pyplot as plt

def predict_prices(dates, prices, x):
    dates = dates.reshape(-1, 1)

    svr_lin = SVR(kernel='linear', C=1e3)
    svr_poly = SVR(kernel='poly', C=1e3, degree=2)
    svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)

    svr_lin.fit(dates, prices)
    svr_poly.fit(dates, prices)
    svr_rbf.fit(dates, prices)

    plt.scatter(dates, prices, color='black')
    plt.plot(dates, svr_rbf.predict(dates), color='red')
    plt.plot(dates, svr_lin.predict(dates), color='green')
    plt.plot(dates, svr_poly.predict(dates), color='blue')
    plt.show()

    return (
        svr_rbf.predict([[x]])[0],
        svr_lin.predict([[x]])[0],
        svr_poly.predict([[x]])[0]
    )

next_day = max(dates) + 1
print(predict_prices(dates, prices, next_day))
```

```python
def predict_prices(dates, prices, x):
    from sklearn.svm import SVR
    from sklearn.preprocessing import StandardScaler
    import numpy as np
    import matplotlib.pyplot as plt

    sc_X = StandardScaler()
    sc_y = StandardScaler()

    X = sc_X.fit_transform(dates.reshape(-1,1))
    y = sc_y.fit_transform(prices.reshape(-1,1)).ravel()

    svr = SVR(kernel='rbf', C=100, gamma=0.01)
    svr.fit(X, y)

    plt.scatter(dates, prices, color='black')
    plt.plot(dates, sc_y.inverse_transform(svr.predict(X).reshape(-1,1)), color='red')
    plt.show()

    pred_scaled = svr.predict(sc_X.transform([[x]]))
    return sc_y.inverse_transform(pred_scaled.reshape(-1,1))[0][0]
```
